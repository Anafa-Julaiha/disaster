 ## **Six Jars of Machine Learning**

- **Data**:
  - **What it is**: The raw information collected for analysis.
  - **Example**: Images, text, numerical data, or any other form of data that can be used to train a machine learning model.

- **Task**:
  - **What it is**: The specific problem you want the machine learning model to solve.
  - **Example**: Classifying emails as spam or not spam, predicting house prices, clustering customers based on purchasing behavior.

- **Model**:
  - **What it is**: The algorithm or method used to learn patterns from the data.
  - **Example**: Decision trees, neural networks, support vector machines.

- **Training**:
  - **What it is**: The process of feeding data into the model to help it learn.
  - **Example**: Providing the model with labeled data so it can learn to classify emails as spam or not spam.

- **Evaluation**:
  - **What it is**: Assessing the model's performance on unseen data to ensure it generalizes well.
  - **Example**: Using a test set to measure accuracy, precision, recall, or other relevant metrics.

- **Deployment**:
  - **What it is**: Integrating the trained model into a real-world application where it can make predictions or decisions.
  - **Example**: Implementing a spam filter in an email service, deploying a house price prediction model in a real estate website.

 Imbalance Issue:

```python
df.describe()
```

```python
df = df.drop(columns=["Time"])
```

```python
scaler = StandardScaler()
df["Amount"] = scaler.fit_transform(df[["Amount"]])
```

```python
# Check split distribution
print("Training Data Fraud Cases:", sum(y_train))
print("Testing Data Fraud Cases:", sum(y_test))
```

## Train & Evaluate Random Forest

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Predictions
y_pred_rf = rf_model.predict(X_test)

# Evaluation
print("Random Forest Classifier Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
```

##  Handle Imbalanced Data (SMOTE)

- Since fraud cases are very few, use SMOTE (Synthetic Minority Over-sampling Technique) to balance the dataset.
-  SMOTE generates synthetic minority class samples to balance the dataset.

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Check class distribution after SMOTE
print("After SMOTE - Class Distribution:\n", pd.Series(y_resampled).value_counts())
```

```python
# Data Preprocessing

# Handling Missing Values

# Fill missing values with median
df.fillna(df.median(), inplace=True)
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['Annual_Income_(k$)', 'Spending_Score']])
```

Why Scale the Data?

- Machine learning models, including K-Means, perform better when features are scaled.
- StandardScaler transforms the data so that it has mean = 0 and standard deviation = 1.
- This ensures all features contribute equally to the clustering process.

## K-Means Clustering

- Find Optimal Number of Clusters (Elbow Method)

```python
wcss = []
for i in range(1, 11):  # Try clusters from 1 to 10
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)  # Inertia = Within-cluster sum of squares
```

```python
# Plot Elbow Method
plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method for Optimal Clusters")
plt.show()
```

What is the Elbow Method?

- K-Means requires us to specify the number of clusters (K) in advance.
- The Elbow Method helps to find the best K by plotting WCSS (Within-Cluster Sum of Squares).
- WCSS measures how tightly grouped the clusters are  lower WCSS means better clustering.
- The "elbow point" (where the curve bends) suggests the optimal number of clusters.

## Train K-Means with Optimal Clusters

```python
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(df_scaled)
```

What Happens Here?

- We set n_clusters=4 based on the Elbow Method.
- The fit_predict() function does two things:
  1. Fits the model (finds cluster centers).
  2. Predicts cluster labels for each data point.
- A new column "Cluster" is added to df, containing the cluster number (0, 1, 2, or 3) assigned to each customer.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Annual_Income_(k$)"], y=df["Spending_Score"], hue=df["Cluster"], palette="viridis")
plt.title("Customer Segments")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```

 What This Visualization Shows:

- Customers are grouped into different colored clusters based on Annual Income vs. Spending Score.
- The visualization helps businesses understand customer behavior:
 - High Income, High Spending  VIP customers
